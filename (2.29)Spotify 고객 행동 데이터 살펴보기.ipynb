```python
# 해당 값 전체 보기
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.options.display.max_columns = None
# 복구
# pd.reset_option('display.max_colwidth')
# pd.reset_option('display.max_rows')
```

```python
# 각 칼럼에 어떤 값들이 있는지 확인
columns = spotify_df.columns
index_df = []
values_df = []
for i in columns:
    index_df.append(i)
    values_df.append(list(spotify_df[i].unique()))
```

```python
column_definition = pd.DataFrame({'index_column': index_df, 'value_column': values_df})
```
